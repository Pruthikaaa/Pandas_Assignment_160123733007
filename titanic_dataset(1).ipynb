iv) AgeGroup with highest fatality rate

```python
bins = [0, 12, 18, 60, 100]
labels = ['Child', 'Teen', 'Adult', 'Senior']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)

fatality_rate = 1 - df.groupby('AgeGroup')['Survived'].mean()
fatality_rate
```

To analyze survival rates across different life stages the 'Age' variable is divided into categorical buckets.
'bins' are used to define intervals/ranges of age, instead of using raw numbers like 25, 30 etc, ages are grouped into categories.
'labels' are names given to each bin so that it becomes easy for us to read and understand.
pd.cut() converts continuous/raw data to categorial data.

Categorical data is way easier to use in groupby() operations. The final output will clearly say if its a 'Child' or a 'Senior' instead of giving confusing intervals.

df.groupby('AgeGroup')['Survived'].mean() will give the 'Survived' rate and 1 - 'Survived' rate will give us 'Fatality' rate.

```python
print(f"The group with the highest fatality rate is: {fatality_rate.idxmax()}")
```

While researching I've found out that idxmax() is used to find the index label of the first occurrence of the maximum value in a Series or DataFrame.

From the output we can clearly see that 'Senior' and 'Adult' age groups have the highest fatality rate.
This can be justified as cruises usually follow 'Women and Children First' policy where they are given access to lifeboats first. 
We can see that the mortality rate for children is less.
On the other hand, the mortality rate for Senior group is more as they were less likely to reach the lifeboats in the chaos.

5) Create a new derived column 'FamilySize'=SibSp + Parch + 1. Using ufuncs, compute a 'SurvivalScore' that weighs Fare and FamilySize. Construct a final ranked DataFrame sorted by SurvivalScore.

i) Creating a new derived column

```python
import numpy as np
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

ii) Using ufuncs and constructing final dataframe

```python
df['SurvivalScore'] = np.log1p(df['Fare']) / df['FamilySize']
ranked_df = df.sort_values(by='SurvivalScore', ascending=False)
ranked_df
```

```python
print(ranked_df[['Name', 'Fare', 'FamilySize', 'SurvivalScore']].head())
```

Simply dividing Fare by FamilySize (Fare-per-person) is a logical way to "weigh" the two variables. However, because Fare has extreme outliers (some paid 0 dollars, others paid 512 dollars), using np.log1p is superior. It squashes the range so that a few wealthy passengers don't completely break the ranking system.

It handles the skewed distribution of fares and accounts for the fact that larger families faced more survival friction.

By using df.sort_values(), a 'Priority List' is created. Final output shows that single passengers in 1st Class appear at the very top, as they had high fares and a family size of 1, resulting in the highest scores.

These operations are vectorized. This means Pandas performs the calculation for all 891 rows simultaneously at the C-level, rather than using a slow Python loop.

6) Perform index alignment between two Series: one containing average fare per Pclass, and anothercontaining average age per Pclass. Examine how Pandas aligns indices during arithmetic operations.

```python
avg_fare = df.groupby('Pclass')['Fare'].mean()
avg_age = df.groupby('Pclass')['Age'].mean()
```

To demonstrate alignment clearly, manually reordering one will show the difference clearly.

```python
addition = avg_fare + avg_age_reordered
subtraction = avg_fare - avg_age_reordered
multiplication = avg_fare * avg_age_reordered
division = avg_fare / avg_age_reordered
```

```python
alignment_check = pd.DataFrame({
    'Avg Fare': avg_fare,
    'Avg Age (Reordered)': avg_age_reordered,
    'Addition': addition,
    'Division': division,
})

print(alignment_check)
```

Pandas does not perform arithmetic operations based on the position of the data (like a list or NumPy array), but based on the index labels.

Even though we reordered avg_age_reordered to start with Pclass 3, Pandas ignores the physical order. 
When avg_fare + avg_age_reordered is executed, it looks for the label 1 in both Series, adds those specific values, and places the result at index 1.
Even when the data is "jumbled", Pandas correctly identifies that the Fare for Pclass 3 should only be added to the Age for Pclass 3.

8) Design a function using apply() and map() to categorize passengers into 'Child', 'Adult', and 'Senior'; based on age. Integrate this with a pivot table showing survival rates per age category and gender.

i) Categorizing passengers

```python
def categorize_age(age):
    if pd.isnull(age):
        return 'Unknown'
    if age < 18:
        return 'Child'
    elif age < 60:
        return 'Adult'
    else:
        return 'Senior'
```

ii) Using apply() and map() functions

```python
df['AgeCategory'] = df['Age'].apply(categorize_age)
```

```python
df['SurvivalStatus'] = df['Survived'].map({0: 'Died', 1: 'Survived'})
```

In Pandas, apply() is typically used for applying a custom function along an axis (like a column), while map() is often used for element-wise transformations or substituting values based on a dictionary/Series.
We use apply() to execute the custom function logic on the 'Age' column.
We use map() to create a human-readable 'Survived' label.

iii) Integrating with a pivot table

```python
pivot_table = df.pivot_table(values='Survived', 
                             index='AgeCategory', 
                             columns='Sex', 
                             aggfunc='mean')

pivot_table
```

From the pivot table, we can see that Adult Females and Children have the highest survival rates, while Adult Males have the lowest. 
This statistical evidence supports the historical priority given to vulnerable groups during the evacuation.

```python
df.isnull().sum()
```

```python
df.dropna(axis=1, thresh=700)
```

We use this for columns like 'Cabin'.
If a column doesn't meet a minimum requirement of "real" data, the entire feature is unreliable for analysis or machine learning. 
Dropping the column is better than guessing 77% of the values.

```python
df.dropna(axis=0, thresh=5)
```

We use this as a safety net for the entire DataFrame.
This ensures that if a specific passenger record is almost entirely empty (e.g., it only has a Name and ID but no Class, Age, or Survival data), it gets removed. 
A threshold of 5 is conservative as it keeps almost everyone except the most broken records.

```python
df_optimal = df.drop(columns=['Cabin'])
df_optimal['Age'] = df_optimal['Age'].fillna(df_optimal['Age'].median())
df_optimal
```

```python
df_final = df_optimal.dropna(subset=['Embarked'])
df_final
```

Based on 'Handling Missing Values' principles, df_optimal and df_final are the most optimal because they follow the three main pillars of data cleaning:

Drop Columns when the data is too sparse to be trusted (Cabin).
Impute values when the data is essential but partially missing (Age).
Drop Rows when the missingness is so small that removal doesn't hurt the sample size (Embarked).
